```python
# Spark provides robust set of pre-defined functions as part of pyspark.sql.functions.

# However, they might not fulfill all our requirements.

# At times, we might have to develop custom UDFs for these scenarios.

# No function available for our requirement while applying row level transformations.

# Also, we might have to use multiple functions due to which readability of the code is compromised.
```

```python
# Here are the steps we need to follow to develop and use Spark User Defined Functions.

# Develop required logic using Python as programming language.

# Register the function using spark.udf.register. Also assign it to a variable.

# Variable can be used as part of Data Frame APIs such as select, filter, etc.

# When we register, we register with a name. That name can be used as part of selectExpr or as part of Spark SQL queries using spark.sql.
```

```python

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('SparkByExamples.com').getOrCreate()


help(spark.udf)
```

```python
df = spark.read.json("orders")
```

```python
df.show(4)
```

```python
## Registering the function



dc = spark.udf.register('date_convert', lambda d: int(d[:10].replace('-', '')))

# using the UDFs in Dataframe API methods

df.select(dc("order_date").alias("order_date")).show(4)
```

```python
type(dc)
```

```python
# filtering 

# u cant use sql type syntax in filtering

df.filter( dc('order_date') == 20140101).show(4)
```

```python
df.groupBy(dc("order_date").alias("order_date")).count().withColumnRenamed("count","order_count").show()
```

```python
help(df.groupBy)
```

## Using UDF in SelectExpr

```python

# u have to use function name( date_convert, registered name ) rather than  dc  in SPARK-SQL and selectExpr

df.selectExpr("date_convert(order_date) AS order_date").show(4)
```

```python
df.selectExpr("order_status").show(4)
```

## Using UDF in spark SQL

```python
## To use udf in spark sql, we need to create a table using createOrReplaceTempView
```

```python
df.createOrReplaceTempView("orders")
```

```python
# Create a UDF

date_convert = spark.udf.register(lambda d : d)
```

```python
spark.sql("""

    SELECT *, date_convert(order_date)
    FROM orders

""").show(4)
```

```python
spark.sql('''
    SELECT date_convert(order_date) AS order_date, count(*) AS order_count
    FROM orders AS o
    GROUP BY 1
'''). \
    show(4)
```

```python
data_cleaning = spark.udf.register("data_cleanse", data_cleanse)
```

```python

# In, dataframe Apis, u have to use the spark register declration varaiable ( data_cleaning )

from pyspark.sql.functions import col

# courses_df.select(data_cleanse(col("course_id"))
#                     ).show()




courses_df.select(
    data_cleaning(col('course_id')).alias('course_id'),
    data_cleaning(col('course_status').alias('course_status'))
).show()
     
```

```python
# using spark sql

# in spark_sql u have to use spark udf_registerd_name( data_cleanse )

courses_df.createOrReplaceTempView('courses')

spark.sql("""

    SELECT course_id, data_cleanse(course_status) AS course_status
    FROM courses

""").show(4)
```
