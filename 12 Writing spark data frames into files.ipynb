## Writing Spark DataFrames into CSV

```python
# We can write the data from Spark Data Frame into CSV files using multiple approaches.

# Approach 1: df.write.csv('path_to_folder')

# Approach 2: df.write.format('csv').save('path_to_folder')

# The column names from the schema can be added as header to each of the files by saying header=True.

# We can also write the data into files using characters other than comma as delimiters or separators.

# We can also compress the data while writing into files using csv.
```

```python

# Default behavior
# It will delimit the data using comma as separator

courses_df.write.csv(f'/Users/{username}/SPARK/courses/csv', mode='overwrite')

# or 

courses_df.write.format("csv").save(f"/Users/{username}/SPARK/courses/csv",mode="overwrite")
```

```python

orders. \
    coalesce(1). \
    write. \
    mode('overwrite'). \
    csv(f'{output_dir}/orders', sep='|', header=True, compression='gzip')

# or

# with option

# orders. \
#     coalesce(1). \
#     write. \
#     mode('overwrite'). \
#     option('compression', 'gzip'). \
#     option('header', True). \
#     option('sep', '|'). \
#     csv(f'{output_dir}/orders')


# we can not use mode with any option


# Another method

orders. \
    coalesce(1). \
    write. \
    mode('overwrite'). \
    format("csv").\
    save(f'{output_dir}/orders', sep='|', header=True, compression='snappy')
```

# with options

```python

# # we cannot use mode with options or use mode as parameter in csv 

orders. \
    coalesce(1). \
    write. \
    mode('overwrite'). \
    options(sep='|', header=True, compression='gzip'). \
    csv(f'{output_dir}/orders')

# Another method

options = {
    'sep': '|',
    'header': True,
    'compression': 'snappy'
}
     

orders. \
    coalesce(1). \
    write. \
    mode('overwrite'). \
    options(**options). \
    csv(f'{output_dir}/orders')
     
    
    
```

## Coalesce and Repartitioning the data frame

```python
# coalesce and repartition are functions on top of the dataframe. Do not get confused between coalesce on Data Frame and the coalesce function available to deal with null values in a given column.

# coalesce is typically used to reduce number of partitions to deal with as part of downstream processing.

# repartition is used to reshuffle the data to higher or lower number of partitions to deal with as part of downstream partitioning.

# Make sure to use a cluster with higher configuration, if you would like to run and experience by your self.

# 2 to 3 worker nodes using Standard with 14 to 16 GB RAM and 4 cores each.

# coalesce is used at the starting of operation while repartition is used somewhere in middle

```

```python
help(orders.coalesce)
```

```python

help(df.repartition)
```

```python
# repartition incurs shuffling and it takes time as data has to be shuffled to newer number of partitions.
# Also you can repartition the Data Frame based on specified columns.
# coalesce does not incur shuffling.
# We use coalesce quite often before writing the data to fewer number of files.
```

```python
# Both repartition and coalesce are used in changeing the number of partitions in dataframe/RDD

# coalesce is used in reducing the partitions.It doesnot do shuffle.

# repartition is used in increasing/decreasing the partitions.It shuffles the data between the executors.
```

```python
## Repartitions
```

```python
df = spark.read.json("/Users/manideep/SPARK/orders")
```

```python
df.rdd.getNumPartitions()
```

```python
df.coalesce(16).rdd.getNumPartitions()

# coalesce can not increase the partitons
```

```python

# incurs shuffling

# Watch the execution time and compare with coalesce

df.repartition(106).rdd.getNumPartitions()
```

```python
df.repartition(1000086,"order_id","order_status").rdd.getNumPartitions()

# if we have 186 or more combinations of above 2 columns, 
#                then we can get 186 partitions
```

```python
## Coalesce 

# one of the common usage of coalesce is to write to lesser number of files
```

```python
df.repartition(16).write.mode("overwrite").json("/Users/manideep/SPARK/repartitions",compression="gzip")

os.listdir("/Users/manideep/SPARK/coalesce")

# we see 16 files
# but it is slower as shuffle happens b/w the executors
```
