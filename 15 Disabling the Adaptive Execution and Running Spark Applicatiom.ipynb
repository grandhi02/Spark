## Disabling the adaptive execution

```python


spark.conf.set("spark.sql.adaptive.enabled","false")
```

```python
spark.read.json("orders").groupBy("order_status","order_customer_id").count().\
    write.\
    parquet("/Users/manideep/parquet/",mode="overwrite")
```

```python
## Now we disabled the adaptive execution

## Then the spark.sql.shuffle.partitions is activated as adaptive exectuion is disabled.

## Now the supposed partitions will be 180 as the shuffle is involved .


## If u run above cell, u will see 200 tasks
```

```python
df = spark.read.parquet("/Users/manideep/parquet/")

len(df.inputFiles())

# we get 200 files

# when we disable the adaptive execution, we ended in 200 partitions from shuffle.partitions as shuffle is involved.

# By this , we need more time and more resources .For simple data, it will be overkill.
```

## Running spark with Adaptive Execution

```python
spark.conf.set("spark.sql.adaptive.enabled",True)
```

```python
spark.read.json("orders").groupBy("order_status","order_customer_id").count().\
    write.\
    parquet("/Users/manideep/parquet/",mode="overwrite")
```

```python
## In above
    # At First, it determines the size of the file by going to the details
    # secondly, it is processing the data to actually read the data from files and groups by keys and calculates the partitions based on workload 
    # finally, As adaptive execution is enables, it determined that one task is enough to actually to take it off
    
# Adaptive execution determines the no of partitions that are supposed to used based upon workload.
    
# If u go into the folder, u only see one file as there is one task ( only one partition is there )
```
